# PLOTY AND CUFFLINKS
Plotly is an interactive visualization library  
Cufflinks connects plotly with pandas

# Libraries

```python
import pandas as pd
import numpy as np
import plotly
%matplotlib inline
```

```python
# Check version of plotly
from plotly import __version__
print(__version__)
```

```python
# Import cufflink
import cufflinks as cf
```

```python
# Import libraries from plotly offine
from plotly.offline import download_plotlyjs,init_notebook_mode,plot, iplot
```

```python
# Connect the java script to the notebook
init_notebook_mode(connected = True)
```

```python
cf.go_offline()
```

# Data

```python
# Create data frame
df = pd.DataFrame(np.random.rand(100,4), columns = 'A B C D'.split())
df.head()
```

```python
# Create second data frame
df2 = pd.DataFrame({'Category':['A','B','C'],
                   'Values':[32, 43,50]})
df2.head()
```

# Line Plots

```python
import warnings
warnings.simplefilter(action = 'ignore', category = FutureWarning)

# Create line plot
df.iplot()
```

# Scatter Plots

```python
# Create scatter plot
df.iplot(kind = 'scatter', x = 'A', y = 'B',mode = 'markers',size = 15)
```

# Bar Plots

```python
# Create bar plot
df2.iplot(kind = 'bar',x = 'Category')
```

```python
# Call aggregate function on data frame
df.sum().iplot(kind = 'bar')
```

# Box Plots

```python
# Create box plot
df.iplot(kind = 'box')
```

# 3D Surface Plots

```python
# Create data frame
df3 = pd.DataFrame({'x':[1,2,3,4,5], 
                    'y':[10,20,30,20,10], 
                    'z':[500,400,300,200,100]})
```

** # Create surface plot **  
df3.iplot(kind = 'surface')

# Histograms

```python
# Create histogram of column A
df['A'].iplot(kind ='hist', bins = 50)
```

```python
# Create histogram of entire dataframe
df.iplot(kind ='hist')
```

# Spread Plots

```python
# Create plot
df[['A','B']].iplot(kind ='spread')
```

# Bubble Plots

```python
# Create bubble plot
df.iplot(kind ='bubble', x = 'A', y = 'B', size = 'C')
```

# Scatter Matrix Plots

```python
# Create scatter matrix
df.scatter_matrix()
```
